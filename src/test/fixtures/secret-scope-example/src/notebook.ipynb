# Config notebook

Reads workspace configuration values from a secret scope.

## Secrets

```python
# Secrets with get
workspace_name = dbutils.secrets.get(scope = "jdbc-test", key = "workspace-name")
resource_group = dbutils.secrets.get(scope = "jdbc-test", key = "resource-group")
subscription_id = dbutils.secrets.get(
    scope = "jdbc-test",
    key = "subscription-id"
)
```

```python
# Secrets with getbytes
cert_bytes = dbutils.secrets.getBytes(scope = "jdbc-test", key = "tls-cert")
```
